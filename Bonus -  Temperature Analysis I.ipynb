# Temperature Analysis I

Statistical analysis

Hawaii is reputed to enjoy mild weather all year. Is there a meaningful difference between the temperature 
in, for example, June and December?

```python
#Set dependencies and set-up

from scipy import stats
from scipy import mean
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')

#Import dependencies - Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#Create engine
engine = create_engine(f'sqlite:///../sql-alchemy-challenge/Resources/hawaii.sqlite')

# Declare a Base using `automap_base()`
Base = automap_base()

# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)

# Save References to Each Table
Measurement = Base.classes.measurement
Station = Base.classes.station

# Create/Start a session to query the database
session = Session(engine)
```

```python
# Identify the temperature in June at all stations across all available 
# years in the dataset.

june_temp_results = session.query(Measurement.date, Measurement.tobs ).\
    filter(func.strftime('%m', Measurement.date) == '06').all()

# Take query results and put into a DataFrame
june_temp_results_df = pd.DataFrame(june_temp_results)
june_temp_results_df 
```

```python
# Identify the temperature in December at all stations across all available 
# years in the dataset.

dec_temp_results = session.query(Measurement.date, Measurement.tobs ).\
    filter(func.strftime('%m', Measurement.date) == '12').all()

# Take query results and put into a DataFrame
dec_temp_results_df = pd.DataFrame(dec_temp_results)
dec_temp_results_df 
```

```python
# Mean Values for tobs in June and December
june_mean = june_temp_results_df['tobs'].mean()
december_mean = dec_temp_results_df['tobs'].mean()

print(f'The mean value for Temperature observations for June is {round(june_mean,2)}')
print(f'The mean value for Temperature observations for December is {round(december_mean,2)}')
```

```python
#Setting population variables
jun_pop = june_temp_results_df['tobs']
dec_pop = dec_temp_results_df['tobs']

#Subplot figure size
plt.subplots(figsize=(10,10))

# Scatter Plot of Data
plt.subplot(2, 1, 1)
plt.scatter(range(len(jun_pop)), jun_pop, color='#FF2700', label="June")
plt.scatter(range(len(dec_pop)), dec_pop, color='#0F95D7', label="December")
plt.title('Scatter plot and histogram Temperature Observations for June and December')
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(jun_pop, 10, density=True, alpha=0.7, color='#FF2700', label="June")
plt.hist(dec_pop, 10, density=True, alpha=0.7, color='#0F95D7', label="December")
plt.axvline(jun_pop.mean(), color='#FF2700', linestyle='solid', linewidth=2)
plt.axvline(dec_pop.mean(), color='#0F95D7', linestyle='solid', linewidth=2)
plt.legend()  


#saving and showing figure
plt.savefig(f'../sql-alchemy-challenge/Output_data/Bonus_TempAnI_tobs_JunDec_sctrplt.png')
plt.show()
```

Conducting an Unpaired (independent) t-test

This test is being done on two independent sets of data.

Alternative Hypothesis: There is a meaningful and statistically significant difference in temperature between June and December

Null: Hypothesis: There is no meaningful and statistically significant difference in temperature between June and December


```python
# Unpaired (independent) t-test
stats.ttest_ind(june_temp_results_df['tobs'], dec_temp_results_df['tobs'], equal_var=False)
```

Assuming a 5% significant level, we reject the Null Hypothesis. 

We accept the Alternative Hypothesis that there is a meaningful and statistically significant difference in temperature between June and December
