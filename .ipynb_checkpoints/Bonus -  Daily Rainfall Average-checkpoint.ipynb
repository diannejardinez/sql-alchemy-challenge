# Daily Rainfall Average

```python
#Set dependencies and set-up
#Import Matplot lib
import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

#Import datetime
import datetime as dt

#Import Pandas
import pandas as pd

#Import dependencies - Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#Create engine
engine = create_engine(f'sqlite:///../sql-alchemy-challenge/Resources/hawaii.sqlite')

#Declare a Base using `automap_base()`
Base = automap_base()

#Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)

#Print all of the classes mapped to the Base
Base.classes.keys()

#Save References to Each Table
Measurement = Base.classes.measurement
Station = Base.classes.station

#Create/Start a session to query the database
session = Session(engine)
```

```python
#Calculate the rainfall per weather station using the previous year's matching dates 
#from start trip dates below

#Trip date: 09-01-2011

#Query for the trip date from the previous year
#year ago from trip: 09-01-2010
year_ago_trip = dt.date(2011, 9, 1) - dt.timedelta(days=365)

#Query for trip dates starting from 09-01-2010
last_trip_date = year_ago_trip + dt.timedelta(days=6)

#Query for  dates between 09-01-2010 to 09-07-2010 in measurement table
trip_0910 = session.query(Measurement.station,Measurement.date, Measurement.prcp, ).\
    group_by(Measurement.station).\
    filter(Measurement.date >= year_ago_trip).\
    filter(Measurement.date <= last_trip_date).all()

#Save the Query Results as a Pandas DataFrame and 
#Set the Index to the Date Column & Sort the Dataframe Values by date
trip_0910_df = pd.DataFrame(trip_0910)
trip_0910_df = trip_0910_df.dropna(how='any')
trip_0910_df = trip_0910_df.rename(columns ={'prcp': 'precipitation'})
trip_0910_df = trip_0910_df.drop(['date'], axis=1)
trip_0910_df 
```

```python
#Calculate the daily normals. Normals are the averages for min, avg, and max temperatures

# Query that will calculate daily normals function to 
# calculate min, max, avg temperatures for certain dates


def daily_normals(date):
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(Measurement.date, *sel).filter(func.strftime("%Y-%m-%d", Measurement.date) == date).all()

#range of trip date of 09-01 to 09-07 in 2011 to ensure consecutive days in that year are selected
day1= daily_normals('2011-09-01')
day2= daily_normals('2011-09-02')
day3= daily_normals('2011-09-03')
day4= daily_normals('2011-09-04')
day5= daily_normals('2011-09-05')
day6= daily_normals('2011-09-06')
day7= daily_normals('2011-09-07')

```

```python
# DataFrame function for creating a DataFrame from all of the querries above

def dailynorms_dataframe(day_results):
    daily_normals_df = pd.DataFrame(day_results, columns=['Date','Minimum Daily Temperature', 'Average Daily Temperature', 'Maximum Daily Temperature'])
    return daily_normals_df 

df1 = dailynorms_dataframe(day1)
df2 = dailynorms_dataframe(day2)
df3 = dailynorms_dataframe(day3)
df4 = dailynorms_dataframe(day4)
df5 = dailynorms_dataframe(day5)
df6 = dailynorms_dataframe(day6)
df7 = dailynorms_dataframe(day7)
```

```python
# Merging DataFrame function for merging all DataFrames from above
def merged_dataframes(dframe1 , dframe2):
    daily_normals_df_merged = pd.merge(dframe1, dframe2, how = 'outer')
    return daily_normals_df_merged

daily_normals_df_merged = merged_dataframes(df1 , df2)
daily_normals_df_merged = merged_dataframes(daily_normals_df_merged , df3)
daily_normals_df_merged = merged_dataframes(daily_normals_df_merged , df4)
daily_normals_df_merged = merged_dataframes(daily_normals_df_merged , df5)
daily_normals_df_merged = merged_dataframes(daily_normals_df_merged , df6)
daily_normals_df_merged = merged_dataframes(daily_normals_df_merged , df7)

daily_normals_df_merged = daily_normals_df_merged.set_index('Date')
daily_normals_df_merged 
```

```python
#Creating bar plot with pandas
daily_temp_areaplot = daily_normals_df_merged.plot(kind='area', stacked=False, figsize=(10,5))

#Creating chart title and labels
plt.title("Daily Temperature from 09/01/11 to 09/07/11", fontsize='15')
plt.xlabel("Date", fontsize='15')
plt.xticks(rotation=45)
plt.ylabel("Daily Temperature", fontsize='15')

#saving and showing figure
plt.savefig(f'../sql-alchemy-challenge/Output_data/Bonus_RainfallAve_dailytemp_areaplt.png')


plt.show()
```
